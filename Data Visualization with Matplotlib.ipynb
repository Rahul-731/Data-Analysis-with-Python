```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
```

```python
data = pd.read_csv('C:\\Users\\rk705\\Downloads\\household Task 3.csv')
```

```python
display(data.head(10))
```

```python
#scatter plot with year agains own
plt.scatter(data['year'], data['own'])

#adding title to the scatter plot
plt.title("scatter plot")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
#line chart with year against sum
plt.plot(data['year'])
plt.plot(data['own'])

#adding title to the plot
plt.title("Line chart")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
# Bar chart
plt.bar(data['year'], data['own'])

plt.title("Bar chart")

#setting the x and y labels
plt.xlabel('year')
plt.ylabel('own')

#adding the legends
plt.show()
```

```python
# Histogram
plt.hist(data['income'])

plt.title("Histogram")

#adding the legends
plt.show()
```
